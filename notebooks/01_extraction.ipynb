# Data Extraction
**Dataset:** Global Terrorism Database (GTD) 1970–2017  
**Goal:** Load the raw data and perform initial exploration.

### Imports
Loads `pandas` for data manipulation, `numpy` for numerical operations, and suppresses non-critical warnings to keep output clean.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## 1. Load Data

### Read CSV
Reads the raw GTD CSV file using `latin-1` encoding (required because the file contains special characters) and `low_memory=False` to avoid mixed-type warnings on large files. Prints the total number of rows and columns.

```python
RAW_PATH = '../data/raw/globalterrorismdb_0718dist.csv'

df = pd.read_csv(RAW_PATH, encoding='latin-1', low_memory=False)
print(f'Rows: {df.shape[0]:,}  |  Columns: {df.shape[1]}')
```

### Tail

## 3. Schema & Data Types

### DataFrame Info
Prints a concise summary of the DataFrame including column names, non-null counts, and inferred data types. Useful for spotting columns with many nulls or unexpected types at a glance.

```python
df.info()
```

## 4. Statistical Summary

### Numeric Describe
Generates descriptive statistics (count, mean, std, min, quartiles, max) for all numeric columns. Helps identify the range and distribution of key fields like year, casualties, and coordinates.

```python
df.describe()
```

### Categorical Describe
Generates descriptive statistics for object (string) columns — shows count, number of unique values, the most frequent value, and its frequency. Useful for understanding categorical fields like country, attack type, and group name.

```python
df.describe(include='object')
```

## 5. Missing Values

### Missing Value Analysis
Calculates the total count and percentage of missing values per column, then filters to show only columns that have at least one null. Results are sorted by missing percentage (descending) to highlight the most incomplete columns first.

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({'missing_count': missing, 'missing_%': missing_pct})
missing_df[missing_df['missing_count'] > 0].sort_values('missing_%', ascending=False)
```

Counts the number of terrorist incidents recorded for each year and sorts them chronologically. Shows how the frequency of attacks has changed over the 1970–2017 period.

```python
df['iyear'].value_counts().sort_index()
```

### Top 10 Countries by Attacks
Lists the 10 countries with the highest number of recorded terrorist incidents. Highlights the geographic hotspots in the dataset.

### Top 10 Regions by Attacks
Lists the 10 world regions with the most incidents. Provides a broader geographic view compared to the country-level breakdown.

### Attack Type Distribution
Shows the frequency of each attack type (e.g. Bombing, Armed Assault, Assassination). Reveals which methods are most commonly used by terrorist groups.

```python
df['attacktype1_txt'].value_counts()
```

### Top 10 Terrorist Groups
Lists the 10 most active terrorist groups by number of claimed or attributed attacks in the dataset.

```python
df['gname'].value_counts().head(10)
```

### Casualties Overview
Computes the total, average, and maximum values for kills (`nkill`) and wounded (`nwound`) across all incidents. Gives a high-level picture of the human impact of the attacks.

```python
df[['nkill', 'nwound']].agg(['sum', 'mean', 'max']).round(2)
```
